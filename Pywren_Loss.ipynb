```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
datas = []

files = ['cirrus.csv']
for i in os.listdir('.'):
    if 'lr' in i:
        files.append(i)
files = sorted(files)
print files

for fname in files:
    datas.append(np.genfromtxt(fname, delimiter=',', names=['x', 'y']))
    


```

```python
f = plt.figure(figsize=(20, 20))
for i in range(len(datas)):
    #a = f.add_subplot(6, 1, i+1)
    #a.plot(datas[i]['x'], -datas[i]['y'], color='r', label='the data')
    if i == 0:
        plt.plot(datas[i]['x'], datas[i]['y'], label=files[i])
    else:
        plt.plot(datas[i]['x'], -datas[i]['y'], label=files[i])
plt.legend()
plt.show()
```
